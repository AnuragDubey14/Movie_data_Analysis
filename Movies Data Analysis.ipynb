### Identify the top three genres that have gained popularity in the last five years

```python
recent_years = range(df['year_of_release'].max() - 4, df['year_of_release'].max() + 1)
df_recent = df[df['year_of_release'].isin(recent_years)]

# Count the occurrences of each genre in each recent year
genre_counts_recent = df_recent.groupby(['year_of_release', 'Genre']).size().unstack().fillna(0)

# Calculate the change in the number of movies for each genre over the years
genre_change = genre_counts_recent.diff(axis=1).sum().sort_values(ascending=False)
top_three_genres=genre_change.head(3)

# Plot a bar chart to visualize the change in the number of movies for each genre
genre_change[top_three_genres.index].plot(kind='bar', colormap='RdBu')
plt.title('Change in Popularity of Top Three Genres in the Last Five Years')
plt.xlabel('Genre')
plt.ylabel('Change in Number of Movies')
plt.show()
```

### Analyze the distribution of movie releases across months and years.

```python
months=df['release_date'].str.split(' ',expand=True)[1]
months.unique()
```

```python

d={'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,'October':10,'November':11,'December':12}
def extract_month(m,d):
    if m in d:
        return d[m]
    else:
        return np.nan
```

### Compare the average IMDb ratings of movies in the Biography and Drama genres.

```python
biography_drama_df = df[df['Genre'].isin(['Biography', 'Drama'])]

plt.figure(figsize=(8, 6))
sns.boxplot(x='Genre', y='imdb_rating', data=biography_drama_df, palette='Set2')
plt.title('Comparison of IMDb Ratings for Biography and Drama Movies')
plt.xlabel('Genre')
plt.ylabel('IMDb Rating')
plt.show()
```

### Perform a statistical test to determine if there is a significant difference in ratings.

```python
from scipy.stats import ttest_ind

# Separate IMDb ratings for Biography and Drama genres
biography_ratings = biography_drama_df[biography_drama_df['Genre'] == 'Biography']['imdb_rating']
drama_ratings = biography_drama_df[biography_drama_df['Genre'] == 'Drama']['imdb_rating']

# Perform an independent two-sample t-test
t_stat, p_value = ttest_ind(biography_ratings, drama_ratings, equal_var=False)

# Print the results
print(f'Test Statistic: {t_stat}')
print(f'P-value: {p_value}')

# Determine significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in ratings.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in ratings.")
```

### Analyze the distribution of movie runtimes (duration) across different genres.

```python
df.runtime=pd.to_numeric(df['runtime'],errors='coerce')
```

```python
genre_avg_runtime = df.groupby('Genre')['runtime'].mean().sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_avg_runtime.values, y=genre_avg_runtime.index, palette='viridis')

plt.title('Average Runtime of Top 10 Genres')
plt.xlabel('Average Runtime (minutes)')
plt.ylabel('Genre')
plt.show()
```

### Identify the genre with the longest and shortest average runtimes.

```python

longest_genre = genre_avg_runtime.idxmax()
shortest_genre = genre_avg_runtime.idxmin()

print(f'The genre with the longest average runtime is: {longest_genre}')
print(f'The genre with the shortest average runtime is: {shortest_genre}')
```

### Create a bar chart to show the average runtime for the top three genres.

```python

# Select the top three genres
top_three_genres = genre_avg_runtime.head(3)

# Create a bar chart
plt.figure(figsize=(10, 6))
ax=sns.barplot(x=top_three_genres.index, y=top_three_genres.values)
ax.bar_label(ax.containers[0])
plt.title('Average Runtime for Top Three Genres')
plt.xlabel('Genre')
plt.ylabel('Average Runtime')
plt.xticks(rotation=45)
plt.show()
```

### Explore the correlation between IMDb ratings, box office performance, and runtime.
   - Visualize the correlation matrix using a heatmap.
   - Identify the factors that have the highest correlation with IMDb ratings.


```python


# Select relevant columns for correlation analysis
correlation_data = df[['imdb_rating', 'imdb_votes', 'runtime']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix: IMDb Ratings, Box Office, and Runtime')
plt.show()

# Identify factors with the highest correlation with IMDb ratings
imdb_rating_correlation = correlation_matrix['imdb_rating'].abs().sort_values(ascending=False)
print("Factors with the highest correlation with IMDb ratings:")
print(imdb_rating_correlation)
```
